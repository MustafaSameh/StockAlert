```python
def duplicates_sentizer():
    global data
    if data.duplicated().sum() != 0 :  
       data = data.drop_duplicates().copy()
    return data.duplicated().sum()
```

```python
duplicates_sentizer()
```

```python
def outliers_sentizer():
    
    q1 = pd.Series(sorted(data["studied_hours"])).quantile(0.25)
    q3 = pd.Series(sorted(data["studied_hours"])).quantile(0.75)
    IQR = q3-q1
    upper_fence = q3 + 1.5 * IQR
    lower_fence = q1 + 1.5 * IQR
    data["studied_hours"] = data["studied_hours"].clip(lower_fence , upper_fence)
```

```python
outliers_sentizer()
```

```python
def nulls_sentizer(): 
       
    if data["studied_hours"].isna().sum() != 0 : 
        data["studied_hours"] = data["studied_hours"].fillna(data["studied_hours"].mean())
    return data.isna().sum()    
        
```
